```python
# Procedure 4: Safe File Operations with Backup

from pathlib import Path
import shutil
from datetime import datetime

student_id = "2025-0509"

base_path = Path.home() / "Documents" / "Panghulan_Activity_5"
backup_dir = base_path / f"backup_{student_id}"
backup_dir.mkdir(exist_ok=True, parents=True)

# Function to write with a backup (Using Surname: Panghulan)
def write_with_backup(filename_str, content):
    filename = Path(filename_str)
    full_path = base_path / filename
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{full_path.stem}_Panghulan_backup_{timestamp}{full_path.suffix}"
    backup_path = backup_dir / backup_name

    # Step 1: Create a backup of the existing file
    if full_path.exists():
        shutil.copy(full_path, backup_path)
        print(f"Backup created: {backup_path.name}")

    # Step 2: Write new content to the original file
    with open(full_path, "w", encoding="utf-8") as file:
        file.write(content)
    print(f"File '{full_path.name}' updated successfully.")

# Step 3: Define function to list specific backups
def list_all_backups():
    print(f"\n--- Panghulan Backups in {backup_dir.name} ---")
    search_pattern = "*Panghulan_backup*"
    for backup in backup_dir.glob(search_pattern):
        print("-", backup.name)

# --- Step 4: Execution Section ---
print("=== File Operations Demo ===")

# 1. Initial creation
print("\n1. Creating new file:")
write_with_backup("demo.txt", "Initial content")

# 2. Update (This will trigger the backup)
print("\n2. Updating file (with backup):")
write_with_backup("demo.txt", "Updated content")

# Step 5: Listing backups
list_all_backups()
```

```python
# Procedure 5: Menu-Driven File Manager

# Step 1: Import required modules
from pathlib import Path
from datetime import datetime
import shutil

# Step 2: Define your custom working directory in Documents 
backup_dir = Path.home() / "Documents" / "PythonFiles" 
backup_dir.mkdir(exist_ok=True)  # Create folder if it doesn’t exist 

# Step 3: Define the menu-driven file manager 
def file_manager(): 
    file_name = input("Enter filename (e.g., notes.txt): ") 
    file_path = backup_dir / file_name 

    while True: 
        print("\n--- MENU ---") 
        print("1. Write to file") 
        print("2. Append to file") 
        print("3. Read file") 
        print("4. Backup files")
        print("5. List all backup files") 
        print("6. Exit") 

        choice = input("Choose an option (1-6): ") 

        if choice == "1": 
            content = input("Enter content to write:\n") 
            with open(file_path, "w", encoding="utf-8") as f: 
               f.write(content) 
            print("File written successfully.") 

        elif choice == "2": 
            more = input("Enter content to append:\n") 
            with open(file_path, "a", encoding="utf-8") as f: 
                f.write("\n" + more) 
            print("Content appended.") 

        elif choice == "3": 
            if file_path.exists(): 
                with open(file_path, "r", encoding="utf-8") as f: 
                    print("\nFile Content:\n", f.read())
            else: 
                print("File not found.") 

        elif choice == "4": 
            if file_path.exists(): 
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S") 
                backup_file = file_path.with_name( 
                    f"{file_path.stem}_backup_{timestamp}{file_path.suffix}" 
                ) 
                shutil.copy2(file_path, backup_file) 
                print(f"Backup created: {backup_file.name}") 
            else: 
                print("Cannot backup. File does not exist.") 

        elif choice == "5": 
            # Try It: Add a menu option to list all backup files
            print(f"\n--- All Backups in {backup_dir.name} ---")
            for backup in backup_dir.glob("*backup*"):
                print("-", backup.name)
        
        elif choice == "6": 
            print("Exiting the file manager.") 
            break 

        else: 
            print("Invalid choice. Please try again.") 
            
# Step 4: Run the menu-driven tool 
if __name__ == "__main__":
    file_manager()
```

```python
# IV. DATA AND OBSERVATION 20

#  Merge Two Files 
from pathlib import Path 
 
# Personal details 
student_id = "2025-0509"
 
# Define file paths 
documents_path = Path.home() / "Documents" / "Panghulan_Activity_5"
f1 = documents_path / f"intro_{student_id}.txt" 
f2 = documents_path / f"lines_{student_id}.txt" 
merged = documents_path / f"merged_{student_id}.txt" 
 
# Ensure both source files exist 
if not f1.exists(): 
    f1.write_text(f"Welcome {student_id} to File Handling in Python!") 
    print(f"Sample intro file created for Student ID: {student_id}") 
 
if not f2.exists(): 
    f2.write_text("Line 1\nLine 2\nLine 3") 
    print(f"Sample lines file created for Student ID: {student_id}") 
 
# Merge contents of both files 
with merged.open("w") as mf: 
    mf.write(f1.read_text()) 
    mf.write("\n") 
    mf.write(f2.read_text()) 
 
print(f"Files merged successfully for Student ID: {student_id}")
```
